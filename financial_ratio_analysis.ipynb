Financial Ratio Analysis Dashboard

## Problem Definition
This project compares financial ratios of six major US companies (Apple, Microsoft, Alphabet, Tesla, Johnson & Johnson, Walmart) from 2023 to 2025. The target users are accounting students learning financial statement analysis and individual investors seeking quick company comparisons.

## Data Source
Data is sourced from WRDS Compustat (Fundamentals Annual table). Access date: April 24, 2026. Variables include: sale (revenue), cogs (cost of goods sold), ni (net income), at (total assets), lt (total liabilities), ceq (common equity).

## Step 1: Import Libraries and Connect to WRDS

```python
import wrds
import pandas as pd

# Connect to WRDS with username (password will still be prompted)
db = wrds.Connection(wrds_username='enxicai2402')
print("Connected to WRDS successfully")
```

## Step 2: Retrieve Data from Compustat

```python
# SQL query to retrieve financial data for 6 companies
query = """
SELECT gvkey, conm, fyear, datadate, sale, cogs, ni, at, lt, ceq
FROM comp.funda
WHERE conm IN ('APPLE INC', 'MICROSOFT CORP', 'ALPHABET INC', 
               'TESLA INC', 'JOHNSON & JOHNSON', 'WALMART INC')
  AND fyear >= 2023
  AND indfmt = 'INDL'
  AND datafmt = 'STD'
  AND consol = 'C'
  AND sale IS NOT NULL
ORDER BY conm, fyear
"""

df = db.raw_sql(query)
print(f"Retrieved {len(df)} records")
df.head()
```

## 


Step 3: Close WRDS Connection

```python
db.close()
print("WRDS connection closed")
```

## Step 4: Calculate Financial Ratios

The following ratios are calculated:
- Gross Margin = (Revenue - COGS) / Revenue
- Net Margin = Net Income / Revenue
- ROE = Net Income / Common Equity
- ROA = Net Income / Total Assets
- Debt Ratio = Total Liabilities / Total Assets

```python
# Calculate financial ratios
df['gross_margin'] = (df['sale'] - df['cogs']) / df['sale']
df['net_margin'] = df['ni'] / df['sale']
df['roe'] = df['ni'] / df['ceq']
df['roa'] = df['ni'] / df['at']
df['debt_ratio'] = df['lt'] / df['at']

# Round to 4 decimal places for readability
for col in ['gross_margin', 'net_margin', 'roe', 'roa', 'debt_ratio']:
    df[col] = df[col].round(4)

print("Financial ratios calculated successfully")
df[['conm', 'fyear', 'gross_margin', 'net_margin', 'roe', 'debt_ratio']].head(10)
```

## Step 5: Save Processed Data to CSV

The calculated ratios are exported to 'financial_ratios.csv' for use in the Streamlit dashboard.

```python
# Save to CSV
df.to_csv('financial_ratios.csv', index=False)
print("Data saved to financial_ratios.csv")

# Verify the file was created
import os
print(f"File location: {os.getcwd()}/financial_ratios.csv")
print(f"File size: {os.path.getsize('financial_ratios.csv')} bytes")
```
